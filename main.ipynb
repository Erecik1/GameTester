```python
import getdata
import convertdata
import gameload
import datapraser
import sys
import time
import pandas as pd

#game = gameload.Game("League of Legends (TM) Client","League of Legends.exe")
game = gameload.Game("League of Legends","LeagueClient.exe")
data = getdata.DataCollector(game)
output,start_time = data.save()
convert = convertdata.ConvertData(output)
match_dict, game_duration = convert.dump_data()
make_data_obj = datapraser.DataProcces(input_data=output, game_api_data=match_dict, game_duration=game_duration)
print(make_data_obj.data_frame_return())
```
